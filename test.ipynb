```python
import os

import os

def print_directory_tree(root_dir, ignore_dirs=None, indent=""):
    if ignore_dirs is None:
        ignore_dirs = {"__pycache__", ".git", ".idea", ".vscode"}  # Thêm các thư mục cần bỏ qua

    for root, dirs, files in os.walk(root_dir):
        level = root.replace(root_dir, "").count(os.sep)
        indent = " " * 4 * level
        print(f"{indent}|-- {os.path.basename(root)}/")
        
        # Lọc các thư mục không quan trọng
        dirs[:] = [d for d in dirs if d not in ignore_dirs]
        
        subindent = " " * 4 * (level + 1)
        for f in files:
            print(f"{subindent}|-- {f}")

# Gọi hàm với đường dẫn đến thư mục lớn của bạn
print_directory_tree("app")
```
